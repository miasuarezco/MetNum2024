```python
from sympy import symbols, I, N, pi, expand
x1, y1, z1 = symbols("x1 y1 z1", real=True)
x = 26 + I*x1
y = 65 + I*y1
z = 23 + I*z1

print(x*y*z)
print(expand(x*y*z))
print(expand((x*y)*z))
print(expand(x*(y*z)))
w=1/(pi + I)
print('w=',w)
w=N(w)
print('w=',w)
```
